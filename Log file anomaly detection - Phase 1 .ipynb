```python
import pandas
filename = 'triallog.csv'
names = ['time','desc']
```

```python
data = pandas.read_csv(filename, names=names,encoding = "ISO-8859-1")
```

```python
data
```

```python
messages=[]
log_types=[]
dates=[]
for i in data.time:
    dates.append((i.split())[1])
for i in data.desc:
    messages.append(((i.split())[0],(i.split())[1]," ".join((i.split())[2:])))
    log_types.append((i.split())[0])
```

```python
import matplotlib.pyplot as plt
plt.plot(log_types)
plt.xlabel('Log entry numbers')
plt.ylabel('Log type')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(15,6)
plt.plot(dates)
plt.xlabel('Log entry number')
plt.ylabel('Date')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(15,6)
plt.plot(dates)
plt.xlabel('Log entry number')
plt.ylabel('Date')
for i in range(445):
    if(log_types[i]=='INFO'):
        plt.scatter(i,dates[i],marker='.',color='skyblue')
    if(log_types[i]=='NOTE'):
        plt.scatter(i,dates[i],marker='.',color='green')
    if(log_types[i]=='WARN'):
        plt.scatter(i,dates[i],marker='*',color='yellow')
plt.show()
```
