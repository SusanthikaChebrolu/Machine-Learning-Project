```python
import pandas
filename = 'logToCsv.csv'
names =  ['SL', 'time', 'level', 'class', 'desc']
```

```python
data = pandas.read_csv(filename, names = names)
```

```python
data.head()
```

```python
messages=[]
log_types=[]
dates=[]
for i in data.time:
    dates.append((i.split())[1])
data['desc'].head()
```

```python
import matplotlib.pyplot as plt
plt.plot(dates)
plt.xlabel('Log entry numbers')
plt.ylabel('Log Date')
plt.show()
```

```python
s1 = data['desc'][0]
i = 2
s2 = data['desc'][1]
cnts = [0, 0]

while(i < len(data['desc'])):
    cnt = 0
    cnts.append(cnt)
    while(s1 == s2):
        cnt += 1
        i += 1
        cnts.append(cnt)
        if (i < len(data['desc'])):
            s2 = data['desc'][i]
        else:
            break
    print(cnt)
    s1 = s2
    if (i < len(data['desc'])):
        s2 = data['desc'][i]
    else:
        break
    i += 1
```

```python
import matplotlib.pyplot as plt
plt.plot(cnts)
plt.xlabel('Log entry numbers')
plt.ylabel('Log Repeats')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(range(250, 386),cnts[250:])
plt.xlabel('Log entry numbers')
plt.ylabel('Log Repetitions')
plt.show()
```
